```python
from lbpqca.lattice.reductions_lattice import LLL, BKZ, babai_nearest_plane
from lbpqca.lattice.attacks import BDD_atack, BDD_atack_recovery_secret, SVP_atack, SVP_atack_recovery_secret
from lbpqca.lattice.reductions_problems import RLWE_to_LWE
from lbpqca.polynomial import add_poly, mull_poly
import numpy as np
```

```python
A=np.array([[ [4, 11, 3], [0,  6, 7]],
 [[14,  5,  0], [1, 12, 6]],
 [ [9,  1,  5],  [8,  7, 11]]],dtype=int)

m, n, k, q = 3, 2, 3, 17

e=np.array([ [0, 1,  0],  [0,  0, -1], [1, 0, 1]],dtype=int)

s=np.array([[2,2,4],[5,1,3]],dtype=int)

tmp=[]

for i in range(m):
    l = e[i].copy()
    for j in range(n):
        l=add_poly(l, mull_poly(A[i,j],s[j], q), q)
    tmp.append(l)

b=np.array(tmp,dtype=int)

print('A=',A,'\nb=',b)

```

```python
Ae, be = RLWE_to_LWE(A,b,q)
Ae=Ae.astype(int)
print('Ae=',Ae,'\nbe=',be)
```

```python
print((Ae@s.flatten()-be)%q)
```

```python
h=SVP_atack(Ae,be,q)
print("Kannan's embedding")
print(h)
```

```python
x=BKZ(h,20)
print('Reduces basis')
print(x)
```

```python
sw1=SVP_atack_recovery_secret(Ae,be,x[0],q)
print("recovered secret:")
print(sw1.reshape((n,k)))
print("true secret:")
print(s)
```
