```python
from lbpqca.lattice.reductions_lattice import LLL, BKZ
import numpy as np
```

```python
B = np.random.randint(-10, 11, size=(20, 20))

while np.linalg.matrix_rank(B) < 20:
    B = np.random.randint(-10, 11, size=(20, 20))

print(B)
```

```python
B_LLL = LLL(B)
print(B_LLL)
```

```python
B_BKZ10 = BKZ(B,10)
print(B_BKZ10)
```

```python
print(np.dot(B_LLL[0], B_LLL[0]))
```

```python
print(np.dot(B_BKZ10[0], B_BKZ10[0]))
```
